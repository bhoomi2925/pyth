```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create data using numpy
days = np.array(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
temperatures = np.random.randint(25, 35, size=7)  # random temps between 25 and 35

# Create a DataFrame with pandas
data = pd.DataFrame({'Day': days, 'Temperature': temperatures})

# Plotting with matplotlib
plt.plot(data['Day'], data['Temperature'], marker='o', linestyle='-', color='blue')
plt.title('Weekly Average Temperature')
plt.xlabel('Day')
plt.ylabel('Temperature (Â°C)')
plt.grid(True)
plt.show()
```
